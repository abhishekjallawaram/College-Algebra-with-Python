# Step 8 - Making a Scatterplot Game

To make the game, you can make a loop that plots a random point and asks the user to input the (x,y) coordinates. Notice the `for` loop that runs three rounds of the game. Run the code, play the game, then you can go on to the next step.

````python
import matplotlib.pyplot as plt
import random

score = 0

xmin = -8
xmax = 8
ymin = -8
ymax = 8

fig, ax = plt.subplots()

for i in range(0,3):
    xpoint = random.randint(xmin, xmax)
    ypoint = random.randint(ymin, ymax)
    x = [xpoint]
    y = [ypoint]
    plt.axis([xmin,xmax,ymin,ymax]) # window size
    plt.plot([xmin,xmax],[0,0],'b') # blue x axis
    plt.plot([0,0],[ymin,ymax], 'b') # blue y axis
    plt.plot(x, y, 'ro')
    print(" ")
    plt.grid() # displays grid lines on graph
    plt.show()
    guess = input("Enter the coordinates of the red point point: \n")
    guess_array = guess.split(",")
    xguess = int(guess_array[0])
    yguess = int(guess_array[1])
    if xguess == xpoint and yguess == ypoint:
        score = score + 1

print("Your score: ", score) # notice this is not in the loop


# Only change code above this line
import math_code_test_b as test
test.step08(score)
````

# Step 13 - Solutions as Coordinates

The `linsolve()` function returns a <i>finite set</i>, and you can convert that "finite set" into (x,y) coordinates. Notice how the code parses the `solution` variable into two separate variables. Just run the code to see how this works.

````python
from sympy import *
x,y = symbols('x y')

# Use variables for each equation
first = x + y
second = x - y

# parse finite set answer as coordinate pair
solution = linsolve([first, second], (x, y))
x_solution = solution.args[0][0]
y_solution = solution.args[0][1]

print("x = ", x_solution)
print("y = ", y_solution)
print(" ")
print("Solution: (",x_solution,",",y_solution,")")


# Just run this code
import math_code_test_b as test
test.step01()
````

# Step 15 - Solve and graph a system

Now you can put it all together: solve a system of equations, graph the system, and plot a point where the two lines intersect. Notice how this code is not like the previous solving equations code or the graphing code or the user input code. Python uses `sympy` to get the (x,y) solution and `numpy` to get the values to graph, so the user inputs nummerical values and the code uses them in two different ways. Think about how you would do this if the user input values for ax + by = c.

````python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np

print("First equation: y = mx + b")
mb_1 = input("Enter m and b, separated by a comma: ")
mb_in1 = mb_1.split(",")
m1 = float(mb_in1[0])
b1 = float(mb_in1[1])

print("Second equation: y = mx + b")
mb_2 = input("Enter m and b, separated by a comma: ")
mb_in2 = mb_2.split(",")
m2 = float(mb_in2[0])
b2 = float(mb_in2[1])

# Solve the system of equations
x,y = symbols('x y')
first = m1*x + b1 - y
second = m2*x + b2 - y
solution = linsolve([first, second], (x, y))
x_solution = round(float(solution.args[0][0]),3)
y_solution = round(float(solution.args[0][1]),3)

# Make sure the window includes the solution
xmin = int(x_solution) - 20
xmax = int(x_solution) + 20
ymin = int(y_solution) - 20
ymax = int(y_solution) + 20
points = 2*(xmax-xmin)

# Define the x values once for the graph
graph_x = np.linspace(xmin,xmax,points)

# Define the y values for the graph
y1 = m1*graph_x + b1
y2 = m2*graph_x + b2

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

# line 1
plt.plot(graph_x, y1)

# line 2
plt.plot(graph_x, y2)

# point
plt.plot([x_solution],[y_solution],'ro')

plt.show()
print(" ")
print("Solution: (", x_solution, ",", y_solution, ")")


# Run this code and test it with different equations
````

- **`theta`:** This is the angle converted from degrees to radians. The conversion is necessary because the NumPy library's trigonometric functions, used in the calculation of the projectile's trajectory, expect angles to be expressed in radians. The formula used for the conversion is `theta = np.radians(angle)`. Radians are a different way of measuring angles based on the radius of a circle, where a full circle is $2\pi$ radians, equivalent to 360 degrees.

#### Calculating Time of Flight

The initial velocity directly influences how far and how high the projectile will travel. In projectile motion, the initial velocity is decomposed into horizontal and vertical components, which are used to calculate the distance traveled and the maximum height reached, respectively. The choice of initial velocity is crucial for overcoming obstacles like the randomized wall in the simulation.

- **`t_flight`:** This variable represents the total time the projectile spends in the air, from the moment of launch until it lands back on the ground. The time of flight is calculated using the formula for the time of flight of a projectile launched at an angle:
  
  $$t = \frac{2 v_0 \sin(\theta)}{g}$$

  In this formula:
  - \(v_0\) is the initial velocity of the projectile.
  - \(\sin(\theta)\) is the sine of the launch angle, which determines the vertical component of the initial velocity.
  - \(g\) is the acceleration due to gravity, acting downward.

  The formula derives from the motion equations under uniform acceleration, specifically considering the vertical motion component and the fact that the projectile starts and ends at the same vertical position. The time of flight is influenced by the initial velocity's vertical component and gravity but is independent of the horizontal component. It is a key parameter for determining how long the projectile will be in motion and influences the maximum horizontal distance it can cover.

### Projectile Motion Equations

The simulation calculates the trajectory of a projectile based on classical physics equations for projectile motion. These equations describe the x (horizontal) and y (vertical) positions of the projectile at any given time `t` after it has been launched.

#### Equations for the Projectile's Position

The position of the projectile at any time `t` can be described by two equations:

- **Horizontal Position (`x`):**
  
  The horizontal position of the projectile is given by the equation:
  
  $$x(t) = v_0 \cos(\theta) \cdot t$$
  
  In this equation, `x(t)` represents the horizontal distance traveled by the projectile from the point of launch at any time `t`.

- **Vertical Position (`y`):**
  
  The vertical position of the projectile is described by the equation:
  
  $$y(t) = v_0 \sin(\theta) \cdot t - \frac{1}{2} g t^2$$
  
  Here, `y(t)` represents the height of the projectile above the launch point at any time `t`.

#### Variables in the Equations

- **`v_0` (Initial Velocity):** This is the speed at which the projectile is launched. It significantly affects how far and how high the projectile will go. The initial velocity is a scalar quantity that is given direction by the launch angle, `theta`.

- **`\theta` (Launch Angle in Radians):** This angle determines the direction of the initial velocity. The cosine of `\theta` gives the proportion of `v_0` contributing to the horizontal velocity, while its sine gives the proportion contributing to the vertical velocity.

- **`g` (Acceleration Due to Gravity):** On Earth, this value is approximately 9.81 m/s² downwards. Gravity affects the vertical motion of the projectile, causing it to decelerate as it rises, stop at its peak, and then accelerate downwards.

- **`t` (Time Since Launch):** This variable represents the duration that has elapsed since the projectile was launched. It is used to calculate the projectile's position at any point in time during its flight.

The combination of these equations and variables allows the simulation to accurately model the path of the projectile through space, considering both the effects of the initial launch conditions and the constant acceleration due to gravity acting on the projectile.

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import random

# Constants
g = 9.81  # Gravity (m/s^2)
angle = 45  # Launch angle in degrees
theta = np.radians(angle)  # Convert angle to radians

# Randomize wall position and height
wall_x = random.uniform(20, 50)  # Wall position (m)
wall_height = random.uniform(5, 30)  # Wall height (m)

# Prompt for initial velocity
v0 = float(input("Enter initial velocity (m/s): "))  # Initial velocity

# Calculate time of flight based on initial velocity and launch angle
t_flight = 2 * v0 * np.sin(theta) / g  # Time of flight

# Setup figure and axes for the plot
fig, ax = plt.subplots()
ax.set_xlim(0, 60)
ax.set_ylim(0, 40)
line, = ax.plot([], [], 'b-', lw=2)  # Line for the projectile
wall_line = ax.plot([wall_x, wall_x], [0, wall_height], 'r-', lw=4)  # Wall

# Initialize the line animation
def init():
    line.set_data([], [])
    return line,

# Animation function - called sequentially
def animate(i):
    # Time increments
    t = np.linspace(0, t_flight, num=100)
    # Equations of motion
    x = v0 * np.cos(theta) * t
    y = v0 * np.sin(theta) * t - 0.5 * g * t ** 2
    # Update the line data
    line.set_data(x[:i], y[:i])
    return line,

# Create the animation
ani = FuncAnimation(fig, animate, frames=len(t), init_func=init, blit=True, interval=20)


plt.xlabel("Distance (m)")
plt.ylabel("Height (m)")
plt.title("Projectile Motion over Wall")
plt.grid(True)
plt.show()

ani.save('animation.mp4', writer='ffmpeg', fps=30)

````
