```python
import matplotlib.pyplot as plt
import numpy as np
import random

def scatter_plot_game_with_axes_and_scoring():
    points = np.random.randint(-20, 20, (7, 2))  
    score = 0
    total_questions = len(points)
    
    print("Format of input: x,y (No spaces)")

    for i, point in enumerate(points, start=1):
        plt.figure(figsize=(15, 15))
        plt.scatter(point[0], point[1], c='red', zorder=5)  # Plot the point above grid lines
        plt.xlim(-20, 10)
        plt.ylim(-20, 10)
        
        # Adding ticks to show numbers on axes
        plt.xticks(np.arange(-20, 21, step=1))
        plt.yticks(np.arange(-20, 21, step=1))
        
        plt.axhline(0, color='black', linewidth=1.5)  # X-axis
        plt.axvline(0, color='black', linewidth=1.5)  # Y-axis
        plt.grid(True, which='both', linestyle='--', linewidth=0.5)  # Grid lines
        plt.title(f"Point {i}: Guess the Coordinates")
        plt.show()
        
        user_guess = input(f"Guess the coordinates of point {i} (x,y): ")
        try:
            guess_x, guess_y = map(int, user_guess.split(','))
            if [guess_x, guess_y] == point.tolist():
                print("Correct!\n")
                score += 1  # Increment score for a correct guess
            else:
                print("Incorrect. Moving to the next point.\n")
        except ValueError:
            print("Invalid format. Please use 'x,y'. Moving to the next point.\n")
        plt.close()

    # Display final score
    print(f"Final Score: {score} out of {total_questions}")

# Call the function to test it
scatter_plot_game_with_axes_and_scoring()
```

```python
import random

def generate_two_step_problem():
    """Generates a two-step algebra problem with a solution."""
    a = 0
    while a == 0:  # Ensure 'a' is not zero for two-step problems
        a = random.randint(-20, 20)
    b = random.randint(-20, 20)
    c = random.randint(-20, 20)
    # Equation: ax + b = c
    return a, b, c, (c - b) / a  # Return coefficients and solution


def generate_one_step_problem():
    """Generates a one-step algebra problem with a solution."""
    a = random.randint(-20, 20)  # Coefficient of x
    c = random.randint(-20, 20)  # Constant term
    # Equation: ax = c
    return a, c, c / a if a != 0 else 0  # Handle division by zero


def algebra_practice_game():
    score = 0
    total_problems = 5
    
    print("Welcome to the Algebra Practice Game!")
    print("Solve each of the following algebra problems:")

    for i in range(total_problems):
        if random.choice([True, False]):  # Randomly choose between one-step and two-step problems
            a, c, solution = generate_one_step_problem()
            user_answer = input(f"Problem {i+1}: Solve for x: \n {a} *x = {c}. \n x = ")
        else:
            a, b, c, solution = generate_two_step_problem()
            user_answer = input(f"Problem {i+1}: Solve for x: \n {a}*x + {b} = {c}. \n x = ")
        
        # Validate and check the user's answer
        try:
            if abs(float(user_answer) - solution) < 0.001:  # Allowing for slight differences due to floating point arithmetic
                print("Correct!\n")
                score += 1
            else:
                print(f"Wrong. The correct answer was {solution:.2f}.")
        except ValueError:
            print("Invalid input. Please enter a numerical answer.\n")
    
    print(f"\nYour final score is {score} out of {total_problems}.")


algebra_practice_game()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def projectile_game():
    wall_x = np.random.randint(20, 50)
    wall_height = np.random.randint(5, 30)
    
    def plot_parabola(a, b, c):
        x = np.linspace(0, 60, 400)
        y = a * x**2 + b * x + c
        plt.plot(x, y, label='Projectile Path')
        plt.plot([wall_x, wall_x], [0, wall_height], 'r-', label='Wall')
        plt.ylim(0, 40)
        plt.xlim(0, 60)
        plt.legend()
        plt.show()
    
    print("Adjust the parabola to clear the wall.")
    a = float(input("Enter a: "))
    b = float(input("Enter b: "))
    c = float(input("Enter c: "))
    
    plot_parabola(a, b, c)

projectile_game()
```
